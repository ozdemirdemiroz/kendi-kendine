```python
import matplotlib.pyplot as plt
import numpy as np
```

x ve y eşit nokta sayılarından oluşmalı 

```python
xpoints=([4,6,5,4,3,2,4])
ypoints=([0,11,12,8,12,11,0])

plt.plot(xpoints,ypoints)
plt.show()
```

x noktalarını belirtmezsek , standart olarak 1,2,3,4,... olarak geliyor . buna göre çizer . mesela zaman olarak düşünülebilir

```python
y_2=([5,7,2,3,1])
plt.plot(y_2)
plt.show()
```

random olarak oluşturduğumuz bir datamız var . zaman olarak (t) buradaki size kadarlık bir zaman arrayimizi oluşturuyoruz . 

```python
from numpy import random
y_3=random.randint(100, size=(30))
#time=np.arange(0,60,2)  
time=np.linspace(0,60,30)  #0 dan 60 a kadar 30 eşit parçaya bölünmüş olarak bir array oluşruracak 
plt.plot(time,y_3)
plt.show()
```

marker eklemek istersek marker='o', marker='x'  falan gibi markerın şekli ile gösteerek belirtiyoruz

```python
plt.plot(time,y_3,marker='o')
plt.show()
```

hem çizgiyi hem rengini hemde marker şeklini değiştirmek için "format string" olayı var 'fmt'. marker|line|color
*--r ile (*) ile marker =* oldu , rengi kırmızı(r) , çizgi şekli ise(--)
* '-'  çizgi
* '--' kesikli çizgi
* ':' noktalı
* '-' noktaçizgi

```python
plt.plot(time,y_3,'*--r')
plt.show()
```

ikinci cizgiyi eklemek için y ekseninde başka bir arrayin olması yeterli . eleman sayısı aynı olmalı 

```python
y_4=random.randint(100, size=(30))
plt.plot(time,y_3,'r')
plt.plot(time,y_4,'--g')
plt.show()
```

başlık , axis lere ve çizgileri isimlendirmek için 

```python
#ikinci çizgi için aynı eleman sayısına sahip bir array daha oluşturuyoruz
y_4=random.randint(100, size=(30))

#çizgilere label ile isim veriyoruz
plt.plot(time,y_3,'r',label="before")
plt.plot(time,y_4,'--g',label="after")

#başlık
plt.title("ölçüm datası")

#axis labeling
plt.xlabel("zaman(sn)")
plt.ylabel("ölçüm")

#legend i göstermek için 
plt.legend()

plt.show()
```

grid ( ızgara) eklemek istersek

```python
y_4=random.randint(100, size=(30))

plt.plot(time,y_3,'r',label="before")
plt.plot(time,y_4,'--g',label="after")

plt.title("ölçüm datası")

plt.xlabel("zaman(sn)")
plt.ylabel("ölçüm")

plt.legend()
plt.grid() #grid eklemek için
plt.show()
```

bunu iki ayrı grafikde göstermek için 

```python
y_4=random.randint(100, size=(30))

plt.subplot(2, 1, 1)    #subplot() ile plotları ikiye bölebiliyruz (iki satır, 1 kolon olacakşekilde bu 1. grafik demek  )
plt.plot(time,y_3,'r',label="before")
plt.legend()
plt.title("ölçüm datası")
plt.xlabel("zaman(sn)")
plt.ylabel("ölçüm")

plt.subplot(2, 1, 2)    #subplot() ile plotları ikiye bölebiliyruz (iki satır, 1 kolon olacakşekilde bu 2. grafik demek  )
plt.plot(time,y_4,'--g',label="after")
plt.legend()
#plt.title("ölçüm datası")  #değerleri ile iç içe girdiğinden bastırmayalım 
plt.xlabel("zaman(sn)")
plt.ylabel("ölçüm")

plt.show()
```

Scatter ve bar  olarak göstermek için 

y_4 array ini bir grafikde plot olarak bir grafikde de dağılım ( scatter)  bir grafikde de bar olarak göstermek için 

```python
y_4=random.randint(100, size=(30))

plt.subplot(3, 1, 1)    
plt.plot(time,y_4,'r',label="plot olarak")
plt.legend()
plt.title("ölçüm datası")
plt.xlabel("zaman(sn)")
plt.ylabel("ölçüm")

plt.subplot(3, 1, 2)   
plt.scatter(time,y_4,label="dağılım olarak") # plt.plot yerine plt.scatter kulanarak dağılım şeklinde gösterebiliriz
plt.legend()
plt.xlabel("zaman(sn)")
plt.ylabel("ölçüm")

plt.subplot(3, 1, 3)   
plt.bar(time,y_4,label="bar",color = 'm') # plt.plot yerine plt.bar kulanarak dağılım bar gösterebiliriz , renk=kırmızı
plt.legend()
plt.xlabel("zaman(sn)")
plt.ylabel("ölçüm")


plt.show()
```
